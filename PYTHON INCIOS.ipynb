##python creator

One of the most well-known and currently used programming languages for the use of artificial intelligence, in addition to being part of the best-known languages in the world of computing is Python, this great tool is due to Guido Van Rossum who during his life in Netherlands it developed an innovative program that aimed to replace the ABC programming language and at the same time be able to interact with the AMOEBA operating system.

![Python crador](http://www.maestrosdelweb.com/images/2010/01/introspectiva-guido-van-rossum.jpg)

All of this goes back to the late 80s and early 90s more precisely in 1989 during the Christmas season when an experienced programmer Guido Van Rossum decided to start a project that he had in mind that would be to give continuity to the programming. ABC language since he himself had worked with that language, this language was well known for its ease of work in addition to learning from it while the user is using it, there was only one drawback for Van Rossum, the hardware that existed during the time La Creation of Python hindered the progress of the project for which it was momentarily stalled.

Once certain Python problems were solved, its creator Guido Van Rossum made the decision to publish the code, it was during the year 1991 on February 20 during its development stage, it was observed that the classes with inheritance and exception handling, functions were present in addition to having different modular types such as str, list, dict, among others; In this first version of Python a system of modules appeared, this was achieved thanks to Modula-3, its creator in his words describes the module as one of the largest Python programming units.
